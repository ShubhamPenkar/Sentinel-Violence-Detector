```python
import os
import random
import numpy as np
import librosa
import soundfile as sf
import joblib
import urllib.request
import tarfile
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import (
    accuracy_score, classification_report,
    confusion_matrix, ConfusionMatrixDisplay, roc_auc_score
)
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
VIOLENT_DIR    = r"C:\Users\Sanjana\OneDrive\Desktop\coding\ALL_ML\Violence_detector\selected_violent"
NONVIOLENT_DIR = r"C:\Users\Sanjana\OneDrive\Desktop\coding\ALL_ML\Violence_detector\selected_nonviolent"
 
# Where to save downloaded speech clips
SPEECH_DIR = r"C:\Users\Sanjana\OneDrive\Desktop\coding\ALL_ML\Violence_detector\speech_nonviolent"
 
# How many speech clips to add (more = better, keep balanced with violent)
N_SPEECH_CLIPS  = 300
 
SR              = 16000
CLIP_DURATION   = 3
BITS            = 10
 
OUTPUT_MODEL    = "violence_model_2.pkl"
OUTPUT_SCALER   = "scaler_2.pkl"
```

```python
def download_commonvoice_clips(out_dir, n_clips=300):
    """
    Downloads a small subset of LibriSpeech (test-clean, ~330MB) and
    slices it into 3-second clips labelled as non-violent.
 
    LibriSpeech test-clean contains clean audiobook readings —
    genuine conversational/speech audio, nothing like angry shouting.
    """
    os.makedirs(out_dir, exist_ok=True)
 
    existing = [f for f in os.listdir(out_dir) if f.endswith(".wav")]
    if len(existing) >= n_clips:
        print(f"✅ Speech clips already exist ({len(existing)} found), skipping download.")
        return
 
    TAR_URL  = "https://www.openslr.org/resources/12/test-clean.tar.gz"
    TAR_PATH = "librispeech_test_clean.tar.gz"
    EXTRACT  = "LibriSpeech"
 
    if not os.path.exists(TAR_PATH):
        print("📥 Downloading LibriSpeech test-clean (~330MB)…")
        urllib.request.urlretrieve(TAR_URL, TAR_PATH)
        print("✅ Download complete.")
 
    if not os.path.exists(EXTRACT):
        print("📂 Extracting…")
        with tarfile.open(TAR_PATH, "r:gz") as tar:
            tar.extractall()
        print("✅ Extraction complete.")
 
    # Collect all .flac files
    flac_files = []
    for root, _, files in os.walk(EXTRACT):
        for f in files:
            if f.endswith(".flac"):
                flac_files.append(os.path.join(root, f))
 
    random.shuffle(flac_files)
 
    segment_len = SR * CLIP_DURATION
    count = 0
 
    print(f"✂️  Slicing speech into {CLIP_DURATION}s clips…")
    for flac_path in flac_files:
        if count >= n_clips:
            break
        try:
            y, sr = librosa.load(flac_path, sr=SR)
            for start in range(0, len(y) - segment_len, segment_len):
                if count >= n_clips:
                    break
                clip = y[start:start + segment_len]
                sf.write(os.path.join(out_dir, f"speech_{count}.wav"), clip, SR)
                count += 1
        except Exception as e:
            print(f"  Skipped {flac_path}: {e}")
 
    print(f"✅ Created {count} speech non-violent clips in '{out_dir}'")
```

```python
def adc_quantize(signal, bits=10):
    max_val = np.max(np.abs(signal))
    if max_val == 0:
        return signal
    signal_norm = signal / max_val
    levels = 2 ** bits
    return np.round(signal_norm * (levels / 2)) / (levels / 2)
 
 
def extract_features(file_path, bits=10):
    """
    Extract 40 features instead of the original 16:
 
    Original (16):  energy, zcr, centroid, 13x MFCC mean
    Added (24 new):
      - MFCC std (13)       — how much MFCCs vary, not just their average
      - spectral rolloff    — freq below which 85% of energy sits
      - spectral bandwidth  — spread of the spectrum
      - spectral contrast   — tonal vs noisy content (7 bands)
      - RMS energy mean+std — loudness profile
      - tempo               — rhythm; violent audio is often more erratic
 
    Total: 3 + 13 + 13 + 1 + 1 + 7 + 2 + 1 = 41 features
    """
    y, sr = librosa.load(file_path, sr=SR)
    y_q   = adc_quantize(y, bits)
 
    features = []
 
    # --- Original 3 ---
    energy   = np.sum(y_q ** 2)
    zcr      = np.mean(librosa.feature.zero_crossing_rate(y_q))
    centroid = np.mean(librosa.feature.spectral_centroid(y=y_q, sr=sr))
    features += [energy, zcr, centroid]
 
    # --- MFCC mean + std (26) ---
    mfcc      = librosa.feature.mfcc(y=y_q, sr=sr, n_mfcc=13)
    features += list(np.mean(mfcc, axis=1))   # 13 means
    features += list(np.std(mfcc,  axis=1))   # 13 stds  ← NEW
 
    # --- Spectral rolloff (1) ---
    rolloff  = np.mean(librosa.feature.spectral_rolloff(y=y_q, sr=sr, roll_percent=0.85))
    features += [rolloff]
 
    # --- Spectral bandwidth (1) ---
    bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=y_q, sr=sr))
    features  += [bandwidth]
 
    # --- Spectral contrast (7) ---
    contrast  = librosa.feature.spectral_contrast(y=y_q, sr=sr)
    features  += list(np.mean(contrast, axis=1))
 
    # --- RMS energy mean + std (2) ---
    rms       = librosa.feature.rms(y=y_q)
    features  += [np.mean(rms), np.std(rms)]
 
    # --- Tempo (1) ---
    tempo, _  = librosa.beat.beat_track(y=y_q, sr=sr)
    features  += [float(tempo)]
 
    return np.array(features)
 
 
FEATURE_COUNT = 41   # update if you change extract_features
```

```python
def build_dataset():
    dataset = []
 
    # Existing violent clips
    for f in os.listdir(VIOLENT_DIR):
        if f.endswith(".wav"):
            dataset.append((os.path.join(VIOLENT_DIR, f), 1))
 
    # Existing non-violent clips (movie ambient)
    for f in os.listdir(NONVIOLENT_DIR):
        if f.endswith(".wav"):
            dataset.append((os.path.join(NONVIOLENT_DIR, f), 0))
 
    # NEW: real speech non-violent clips
    if os.path.exists(SPEECH_DIR):
        for f in os.listdir(SPEECH_DIR):
            if f.endswith(".wav"):
                dataset.append((os.path.join(SPEECH_DIR, f), 0))
 
    random.shuffle(dataset)
    return dataset
 
 
def extract_all(dataset):
    X, y = [], []
    errors = 0
    for path, label in tqdm(dataset, desc="Extracting features"):
        try:
            feats = extract_features(path)
            X.append(feats)
            y.append(label)
        except Exception as e:
            errors += 1
    if errors:
        print(f"⚠️  Skipped {errors} files due to errors.")
    return np.array(X), np.array(y)
 
```

```python
def train_and_evaluate(X_train, X_test, y_train, y_test):
    models = {
        "Logistic Regression": LogisticRegression(max_iter=2000, C=1.0, solver="lbfgs"),
        "Random Forest":       RandomForestClassifier(n_estimators=200, random_state=42),
        "Gradient Boosting":   GradientBoostingClassifier(n_estimators=200, random_state=42),
        "SVM (RBF)":           SVC(kernel="rbf", probability=True, random_state=42),
    }
 
    results = {}
 
    print("\n📊 Model Comparison (test set accuracy + 5-fold CV AUC):")
    print("-" * 60)
 
    for name, m in models.items():
        m.fit(X_train, y_train)
        y_pred = m.predict(X_test)
        y_prob = m.predict_proba(X_test)[:, 1]
 
        acc = accuracy_score(y_test, y_pred)
        auc = roc_auc_score(y_test, y_prob)
 
        # 5-fold cross-validation on training set
        cv_scores = cross_val_score(m, X_train, y_train, cv=5, scoring="roc_auc")
 
        print(f"  {name:25s}  acc={acc:.3f}  auc={auc:.3f}  cv_auc={cv_scores.mean():.3f}±{cv_scores.std():.3f}")
        results[name] = {"model": m, "acc": acc, "auc": auc, "cv_auc": cv_scores.mean()}
 
    # Pick best by CV AUC (more reliable than test accuracy alone)
    best_name = max(results, key=lambda k: results[k]["cv_auc"])
    best      = results[best_name]
 
    print(f"\n🏆 Best model: {best_name}  (cv_auc={best['cv_auc']:.3f})")
    return best["model"], best_name, results
 
 
def plot_results(best_model, best_name, X_test, y_test, results):
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
 
    # Confusion matrix
    y_pred = best_model.predict(X_test)
    cm     = confusion_matrix(y_test, y_pred)
    disp   = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Non-Violent", "Violent"])
    disp.plot(ax=axes[0], colorbar=False)
    axes[0].set_title(f"Confusion Matrix — {best_name}")
 
    # Model comparison bar chart
    names  = list(results.keys())
    scores = [results[n]["cv_auc"] for n in names]
    axes[1].bar(names, scores, color=["#22c55e" if n == best_name else "#6b7280" for n in names])
    axes[1].set_ylabel("CV AUC")
    axes[1].set_title("Model Comparison (5-fold CV AUC)")
    axes[1].set_ylim(0.5, 1.0)
    axes[1].tick_params(axis="x", rotation=20)
 
    plt.tight_layout()
    plt.savefig("training_results.png", dpi=150)
    plt.show()
    print("📈 Plot saved as training_results.png")
 
 
def print_classification_report(model, X_test, y_test):
    y_pred = model.predict(X_test)
    print("\n📋 Detailed Classification Report:")
    print(classification_report(y_test, y_pred, target_names=["Non-Violent", "Violent"]))
```

```python
if __name__ == "__main__":
 
    # 1. Download speech data
    print("=" * 60)
    print("STEP 1: Preparing non-violent speech data")
    print("=" * 60)
    download_commonvoice_clips(SPEECH_DIR, n_clips=N_SPEECH_CLIPS)
 
    # 2. Build dataset
    print("\n" + "=" * 60)
    print("STEP 2: Building dataset")
    print("=" * 60)
    dataset = build_dataset()
 
    violent_count    = sum(1 for _, l in dataset if l == 1)
    nonviolent_count = sum(1 for _, l in dataset if l == 0)
    print(f"  Violent clips:     {violent_count}")
    print(f"  Non-violent clips: {nonviolent_count}  (ambient + speech)")
    print(f"  Total:             {len(dataset)}")
 
    # 3. Extract features
    print("\n" + "=" * 60)
    print("STEP 3: Extracting 41 features per clip")
    print("=" * 60)
    X, y = extract_all(dataset)
    print(f"  Feature matrix: {X.shape}")
 
    # 4. Scale + split
    scaler  = StandardScaler()
    X_scaled = scaler.fit_transform(X)
 
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.2, random_state=42, stratify=y
    )
    print(f"  Train: {len(X_train)}  Test: {len(X_test)}")
 
    # 5. Train & evaluate
    print("\n" + "=" * 60)
    print("STEP 4: Training models")
    print("=" * 60)
    best_model, best_name, results = train_and_evaluate(X_train, X_test, y_train, y_test)
 
    # 6. Detailed report
    print_classification_report(best_model, X_test, y_test)
 
    # 7. Plot
    plot_results(best_model, best_name, X_test, y_test, results)
 
    # 8. Save
    joblib.dump(best_model, OUTPUT_MODEL)
    joblib.dump(scaler,     OUTPUT_SCALER)
    print(f"\n✅ Saved: {OUTPUT_MODEL} + {OUTPUT_SCALER}")
    print(f"   Drop these into your Violence_detector folder and rerun app.py")
```
